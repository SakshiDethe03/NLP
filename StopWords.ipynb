### Stop Words

```python
import spacy

from spacy.lang.en.stop_words import STOP_WORDS

print(STOP_WORDS)
len(STOP_WORDS)
```

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("I am Sakshi and I am learning Natural Language Processing")

for token in doc:
    if token.is_stop:
        print(token)
```

```python
def preprocess(text):
    doc = nlp(text)

    no_stop_words = [token.text for token in doc if not token.is_stop and not token.is_punct]
    return no_stop_words

preprocess("I am Sakshi and I am learning Natural Language Processing, I love to work with the AI/ML")
```
