```python
import pandas as pd

df = pd.read_csv("ecommerce_data.csv", names=["description", "category"], header= None)

print(df.shape)
df.head()
```

```python
df.category.value_counts()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.category.replace("Clothing & Accessories", "Clothing_Accessories", inplace=True)
df.category.unique()
```

```python
df['category'] = "__label__" + df['category'].astype(str)
df.head()
```

```python
df['category_desc'] = "__label__" + df['description']
df.head()
```

```python
df['category_desc'][1]
```

```python
import re

def preprocess(text): 
   text  = re.sub(r'[^\w\s]', ' ', text)
   text = re.sub(r' +', ' ', text)
   return text.strip().lower()
```

```python
preprocess(", Wooden Box Size:- 15x15x8 cm Wooden Hand Crafted and Painted ")
```

```python
df['category_desc'] = df['category_desc'].map(preprocess)
```

```python
df.head(10)
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2)
```

```python
train.shape, test.shape
```

```python
test.head()
```

```python
train.to_csv("ecommerce.train", columns = ["category_desc"], index=False, header=False)
test.to_csv("ecommerce.test", columns=["category_desc"], index=False, header=False)
```

```python
import fasttext

model = fasttext.train_supervised(input="ecommerce.train")
model.test("ecommerce.test")
```
